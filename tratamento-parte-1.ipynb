```python
import pandas as pd

colunas_relevantes = [
    'IDAtracacao',
    'Origem',
    'Destino',
    'Tipo Operação da Carga',
    'Tipo Navegação',
    'Carga Geral Acondicionamento',
    'ConteinerEstado',
    'Natureza da Carga',
    'Sentido'
]

# Ler o arquivo CSV e selecionar apenas as colunas relevantes
Cargadf = pd.read_csv(carga2023, delimiter=';', usecols=colunas_relevantes)

print(Cargadf.info())
```

```python
Cargadf
```

```python
valores_unicos = Cargadf['Carga Geral Acondicionamento'].unique()

# Contar a frequência de cada valor na coluna 'Carga Geral Acondicionamento'
contagem_valores = Cargadf['Carga Geral Acondicionamento'].value_counts()

# Exibir os valores únicos
print("Valores únicos na coluna 'Carga Geral Acondicionamento':")
print(valores_unicos)

# Exibir a contagem de frequência de cada valor
print("\nContagem de frequência na coluna 'Carga Geral Acondicionamento':")
print(contagem_valores)
#90298 dados faltantes; 4% dos dados
```

```python
valores_unicos = Cargadf['ConteinerEstado'].unique()

# Contar a frequência de cada valor na coluna 'Carga Geral Acondicionamento'
contagem_valores = Cargadf['ConteinerEstado'].value_counts()

# Exibir os valores únicos
print("Valores únicos na coluna 'ConteinerEstado':")
print(valores_unicos)

# Exibir a contagem de frequência de cada valor
print("\nContagem de frequência na coluna 'ConteinerEstado':")
print(contagem_valores)
#422169 19% dos dados
```

```python
valores_unicos = Cargadf['Origem'].unique()

contagem_valores = Cargadf['Origem'].value_counts()

print(valores_unicos)

print(contagem_valores)
```

```python
# Contar o número de valores NaN em cada coluna
contagem_nan_por_coluna = Cargadf.isna().sum()

# Exibir a contagem de NaN por coluna
print("Contagem de valores NaN por coluna:")
print(contagem_nan_por_coluna)
```
