```python
import pandas as pd
#Retiradas por redudancia Apelido, Ano, Mes, Porto Atracacao, UF
#Retirada por irrelevancia IDberco,Berco,Flag,
#Excesso de nulos Ncaptania, N do Imo, Nacionalidade do Armador

colunas_relevantes = [
    'IDAtracacao',
    'CDTUP',
    'Data Atracação',
    'Data Chegada',
    'Data Desatracação',
    'Data Início Operação',
    'Data Término Operação',
    'Tipo de Operação',
    'Tipo da Autoridade Portuária',
    'Tipo de Navegação da Atracação',
    'Complexo Portuário',
    'Terminal',
    'Município',
    'SGUF',
    'Região Geográfica'
]
#Provavelmente algumas colunas serão removidas para evitar redundâncias ou se forem consideradas irrelevantes

atracacaodf = pd.read_csv(atracacao2023, delimiter=';', usecols=colunas_relevantes)

print(atracacaodf.info())
```

```python
contagem_nan_por_coluna = atracacaodf.isnull().sum()

print("Contagem de valores NaN por coluna:")
print(contagem_nan_por_coluna)
```

```python
atracacaodf
```

```python
firstmergedf = pd.merge(atracacaodf, cargaunicadf, on ='IDAtracacao', how = 'inner')
```

```python
firstmergedf
```

```python
contagem_nan_por_coluna = firstmergedf.isnull().sum()

print("Contagem de valores NaN por coluna:")
print(contagem_nan_por_coluna)
```

```python
df_clean = firstmergedf.dropna(subset=['Tipo de Navegação da Atracação'])
```

```python
df_clean
```

```python
contagem_nan_por_coluna = df_clean.isnull().sum()

print("Contagem de valores NaN por coluna:")
print(contagem_nan_por_coluna)
```

```python
colunas_data =['Data Atracação', 'Data Chegada', 'Data Desatracação', 'Data Início Operação', 'Data Término Operação']

for coluna in colunas_data:
    df_clean[coluna] = pd.to_datetime(df_clean[coluna], format='%d/%m/%Y %H:%M:%S')
```

```python
print(df_clean.info())
```

```python
#Colunas arrumadas: Data Chegada

Ano_coluna = 'Data Chegada'
anos_unicos = pd.DatetimeIndex(df_clean[Ano_coluna]).year.unique()
print(f"Anos únicos na coluna '{Ano_coluna}': {anos_unicos}")
```

```python
# Selecionar as linhas onde o ano é menor que 2022
linhas_ano_menor_2022 = df_clean[df_clean['Data Chegada'].dt.year < 2022]

# Substituir o ano nessas linhas para 2023
df_clean.loc[linhas_ano_menor_2022.index, 'Data Chegada'] = df_clean.loc[linhas_ano_menor_2022.index, 'Data Chegada'].apply(lambda x: x.replace(year=2023))
```

```python
#Colunas arrumadas: Data Chegada

Ano_coluna = 'Data Chegada'
anos_unicos = pd.DatetimeIndex(df_clean[Ano_coluna]).year.unique()
print(f"Anos únicos na coluna '{Ano_coluna}': {anos_unicos}")
```

```python
import pandas as pd

# Filtrar as linhas onde o ano é igual a 2023 e o mês é menor que 12 a partir da 'Data Chegada'
linhas_ano_2023_mes_menor_12 = df_clean[(df_clean['Data Chegada'].dt.year < 2023) & (df_clean['Data Chegada'].dt.month < 12)]

display(linhas_ano_2023_mes_menor_12)
```

```python
# Filtrar as linhas onde o ano é diferente de 2023 e o mês é menor que 12 na coluna 'Data Chegada'
linhas_a_corrigir = df_clean[(df_clean['Data Chegada'].dt.year != 2023) & (df_clean['Data Chegada'].dt.month < 12) & (df_clean['Data Início Operação'].dt.year == 2023)]

# Corrigir as datas da coluna 'Data Chegada' para 2023 nas linhas selecionadas
df_clean.loc[linhas_a_corrigir.index, 'Data Chegada'] = pd.to_datetime(linhas_a_corrigir['Data Chegada'].dt.strftime('%Y-12-%d'))
```

```python

# Vendo o que sobrou tentando encontrar novos erros, ou se são operações de 2022 de fato.
linhas_ano_2023_mes_menor_12 = df_clean[(df_clean['Data Chegada'].dt.year < 2023) & (df_clean['Data Chegada'].dt.month < 12)]

display(linhas_ano_2023_mes_menor_12)
len(linhas_ano_2023_mes_menor_12)
```

```python
#Para trabalharmos somente com o ano de 2023, vamos só utilizar dados que a data chegada é 2023, salvamos o máximo de linhas possíveis.
import pandas as pd


linhas_ano_menor_2023 = df_clean[df_clean['Data Chegada'].dt.year < 2023]

display(linhas_ano_menor_2023)
len(linhas_ano_menor_2023)
```

```python
df_clean.drop(linhas_ano_menor_2023.index, inplace=True)
```
